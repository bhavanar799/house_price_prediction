```python
import pandas as pd
```

```python
td=pd.read_csv("train.csv")
td.head()
```

```python
test_data=pd.read_csv("test.csv")
test_data.head()
```

```python
print("Training data shape :",td.shape)
print("Test data shape :",test_data.shape)
```

```python
print("\nTrainng data info:")
print(td.info())
```

```python
print(test_data.info())
```

```python
#Fill numnerical missing values with median
for column in td.columns:
    if td[column].dtype!='object':
        td[column].fillna(td[column].median(),inplace=True)
        if column in test_data.columns:
            test_data[column].fillna(test_data[column].median(),inplace=True)
            
```

```python
#Fill categorial missing values with mode
for column in td.columns:
    if td[column].dtype=='object':
        td[column].fillna(td[column].mode()[0],inplace=True)
        if column in test_data.columns:
            test_data[column].fillna(test_data[column].mode()[0],inplace=True)
```

```python
#check if there are any missing values left in the training dataset
print("missing values in training data:",td.isnull().sum().sum())

#check if there are any missing values left in the test data
print("Missing values in test data:",test_data.isnull().sum().sum())
```

```python
y=td['SalePrice']
td.drop('SalePrice', axis=1, inplace=True)
```

```python
td = pd.get_dummies(td)
test_data = pd.get_dummies(test_data)
```

```python
td,test_data=td.align(test_data,join='inner',axis=1)
print("Training data shape", td.shape)
print("test data shape",test_data.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
x_train,x_test,y_train,y_test=train_test_split(td,y,test_size=0.2,random_state=0)
```

```python
model=RandomForestRegressor(n_estimators=100,random_state=0)
model.fit(x_train,y_train)
predictions=model.predict(x_test)
me=mean_absolute_error(y_test,predictions)
me
```

```python
model=RandomForestRegressor(n_estimators=500,random_state=0)
model.fit(x_train,y_train)
predictions=model.predict(x_test)
me=mean_absolute_error(y_test,predictions)
me
```

```python
from sklearn.ensemble import GradientBoostingRegressor

# Define the model
model = GradientBoostingRegressor(random_state=0)

# Fit the model
model.fit(x_train, y_train)

# Make predictions on the validation set and evaluate the model
predictions = model.predict(x_test)
mae = mean_absolute_error(y_test, predictions)
print('Mean Absolute Error:', mae)

# Make predictions on the test data
test_predictions = model.predict(test_data)
```

```python
submission=pd.DataFrame({'Id':test_data['Id'],
                        'SalePrice':test_predictions})
```
